```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# Load the Titanic Dataset from Kaggle
```

```python
train_df = pd.read_csv('train.csv')
```

```python
test_df = pd.read_csv('test.csv')
```

```python
# Data Cleaning
```

```python
print(train_df.info())
print(train_df.describe())
```

```python
# Fill missing Age values with the median Age
```

```python
train_df.loc[:, 'Age'] = train_df['Age'].fillna(1)
```

```python
test_df.loc[:, 'Age'] = test_df['Age'].fillna(1)
```

```python
# Drop the cabin column since it has too many missing values
```

```python
train_df.drop('Cabin', axis=1, inplace=True)
```

```python
test_df.drop('Cabin', axis=1, inplace=True)
```

```python
# Fill missing embarked values with 'S'
```

```python
train_df['Embarked'] = train_df['Embarked'].fillna('S')
```

```python
# Exploratory Data Analysis (EDA)
```

```python
# 1 Survival Rate
```

```python
sns.countplot(x='Survived', data=train_df)
plt.title('Survival Rate')
plt.show()
```

```python
# 2 Age Distribution
```

```python
sns.displot(train_df['Age'], kde=False)
plt.title('Age Distribution')
plt.show()
```

```python
# 3 Class Distribution
```

```python
sns.countplot(x='Pclass', data=train_df)
plt.title('Class Distribution')
plt.show()
```

```python
# 4 Sex Distribution
```

```python
sns.countplot(x='Sex', data=train_df)
plt.title('Sex Distribution')
plt.show()
```

```python
# 5 Embarked Distribution
```

```python
sns.countplot(x='Embarked', data=train_df)
plt.title('Embarked Distribution')
plt.show()
```

```python
# 6 Survival Rate by Class
```

```python
sns.barplot(x='Pclass', y='Survived', data=train_df)
plt.title('Survival Rate by Class')
plt.show()
```

```python
# 7 Survival Rate by Sex
```

```python
sns.barplot(x='Sex', y='Survived', data=train_df)
plt.title('Survival Rate by Sex')
plt.show()
```

```python
# 8 Survival Rate by Embarked
```

```python
sns.barplot(x='Embarked', y='Survived', data=train_df)
plt.title('Survival Rate by Embarked')
plt.show()
```
